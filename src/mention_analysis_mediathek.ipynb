```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
plt.ion()
```

```python
btw17 = {
    'linke': 69,
    'grüne': 67,
    'spd': 153,
    'fdp': 80,
    'cdu_csu': 246,
    'afd': 94,
}

btw21 = {
    'linke': 39,
    'grüne': 118,
    'spd': 206,
    'fdp': 92,
    'cdu_csu': 197,
    'afd': 83,
}
```

```python
party_mentions = pd.read_pickle('../data/mentions/party_mentions_mediathek.pkl')
```

```python
party_mentions['cdu_csu'] = party_mentions['cdu'] + party_mentions['csu']
party_mentions.drop(columns=['cdu', 'csu'], inplace=True)
party_mentions = party_mentions[['linke', 'grüne', 'spd', 'fdp', 'cdu_csu', 'afd']]
```

```python
party_mentions_standardized = party_mentions.div(party_mentions.sum(axis=1), axis=0)

f, ax = plt.subplots(figsize=(9, 8))
f.subplots_adjust(left=0.15, bottom=.13, right=.99, top=.89)
sns.heatmap(party_mentions_standardized, annot=True, linewidths=.5, ax=ax, fmt=".2%", center=np.nanmean(party_mentions_standardized), cbar=False, cmap=sns.diverging_palette(145, 20, as_cmap=True).reversed())
```

```python
party_mentions_17 = party_mentions.copy()
for party in btw17.keys():
    party_mentions_17[party] /= btw17[party]

party_mentions_17_standardized = party_mentions_17.div(party_mentions_17['afd'], axis=0)

f, ax = plt.subplots(figsize=(9, 8))
f.subplots_adjust(left=0.15, bottom=.13, right=.99, top=.89)
sns.heatmap(party_mentions_17_standardized, annot=True, linewidths=.5, ax=ax, fmt="1.2", center=np.nanmean(party_mentions_17_standardized), cbar=False, cmap=sns.diverging_palette(145, 20, as_cmap=True).reversed())
```

```python
party_mentions_21 = party_mentions.copy()
for party in btw21.keys():
    party_mentions_21[party] /= btw21[party]

party_mentions_21_standardized = party_mentions_21.div(party_mentions_21['afd'], axis=0)

f, ax = plt.subplots(figsize=(9, 8))
f.subplots_adjust(left=0.15, bottom=.13, right=.99, top=.89)
sns.heatmap(party_mentions_21_standardized, annot=True, linewidths=.5, ax=ax, fmt="1.2", center=np.nanmean(party_mentions_21_standardized), cbar=False, cmap=sns.diverging_palette(145, 20, as_cmap=True).reversed())
```
