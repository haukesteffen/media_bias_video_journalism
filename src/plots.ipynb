```python
from pathlib import Path
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
mpl.use('pdf')

fpath = Path(mpl.get_data_path(), "fonts/ttf/cmu_serif_roman.ttf")
fprop = fm.FontProperties(fname=fpath)

width = 6.2 #inches
height = width

font = {'family':'serif', 'serif': ['cmr10'], 'size':12}
plt.rc('font',**font)
plt.rcParams['axes.unicode_minus'] = False
plt.rc('axes', titlesize=12)
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)

def saveplot_sentiment(df, title, path):
    f, ax = plt.subplots(figsize=(9, 8))
    f.subplots_adjust(left=0.15, bottom=.08, right=.99, top=.93)
    sns.heatmap(df, annot=True, linewidths=.5, ax=ax, fmt=".1%", center=np.nanmean(df), cbar=False)

    ax.set_xlabel('Partei', fontsize=12, font=fpath, color='black')
    ax.set_ylabel('Medium', fontsize=12, font=fpath, color='black')
    ax.set_title(title, fontsize=12, font=fpath, color='black')
    ax.set_xticklabels(['Linke', 'Grüne', 'SPD', 'FDP', 'CDU', 'CSU', 'AfD'], size=12, font=fpath, color='black')
    ax.set_yticklabels(['JW', 'NDS', 'taz', 'SZ', 'St', 'Sp', 'ZEIT', 'TS', 'ARD', 'ZDF', 'ZDFh', 'BR', 'ntv', 'RTL', 'FOCUS', 'faz', 'WELT', 'BILD', 'NZZ', 'JF', 'CTV'], size=12, font=fpath, color='gray')
    
    plt.gca().get_yticklabels()[1].set_color('black') 
    plt.gca().get_yticklabels()[2].set_color('black') 
    plt.gca().get_yticklabels()[5].set_color('black') 
    plt.gca().get_yticklabels()[7].set_color('black') 
    plt.gca().get_yticklabels()[8].set_color('black') 
    plt.gca().get_yticklabels()[9].set_color('black') 
    plt.gca().get_yticklabels()[10].set_color('black') 
    plt.gca().get_yticklabels()[11].set_color('black') 
    plt.gca().get_yticklabels()[13].set_color('black') 
    plt.gca().get_yticklabels()[15].set_color('black') 
    plt.gca().get_yticklabels()[16].set_color('black') 
    plt.gca().get_yticklabels()[17].set_color('black') 
    plt.gca().get_yticklabels()[20].set_color('black') 
    
    f.set_size_inches(width, height)
    f.savefig(path)
```
