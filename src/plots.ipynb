```python
from pathlib import Path
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
mpl.use('pdf')

fpath = Path(mpl.get_data_path(), "fonts/ttf/cmu_serif_roman.ttf")
fprop = fm.FontProperties(fname=fpath)

width = 6.2 #inches
height = width

font = {'family':'serif', 'serif': ['cmr10'], 'size':12}
plt.rc('font',**font)
plt.rcParams['axes.unicode_minus'] = False
plt.rc('axes', titlesize=12)
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)

def saveplot_sentiment(df, title, path, n_rows_in_title=1, lang='de'):
    f, ax = plt.subplots(figsize=(9, 8))
    if n_rows_in_title==1:
        f.subplots_adjust(left=0.15, bottom=.08, right=.99, top=.96)
    elif n_rows_in_title==2:
        f.subplots_adjust(left=0.15, bottom=.08, right=.99, top=.93)

    sns.heatmap(df, annot=True, linewidths=.5, ax=ax, fmt=".1%", center=np.nanmean(df), cbar=False, cmap=sns.diverging_palette(145, 20, as_cmap=True).reversed())

    if lang=='de':
        ax.set_xlabel('Partei', fontsize=12, font=fpath, color='black')
        ax.set_ylabel('Medium', fontsize=12, font=fpath, color='black')
    elif lang=='en':
        ax.set_xlabel('Party', fontsize=12, font=fpath, color='black')
        ax.set_ylabel('Medium', fontsize=12, font=fpath, color='black')
    ax.set_title(title, fontsize=12, font=fpath, color='black')
    ax.set_xticklabels(['Linke', 'Grüne', 'SPD', 'FDP', 'CDU', 'CSU', 'AfD'], size=12, font=fpath, color='black')
    ax.set_yticklabels(['JW', 'NDS', 'taz', 'SZ', 'St', 'Sp', 'ZEIT', 'TS', 'ARD', 'ZDF', 'ZDFh', 'BR', 'ntv', 'RTL', 'FOCUS', 'faz', 'WELT', 'BILD', 'NZZ', 'JF', 'CTV'], size=12, font=fpath, color='gray')
    
    plt.gca().get_yticklabels()[1].set_color('black') 
    plt.gca().get_yticklabels()[2].set_color('black') 
    plt.gca().get_yticklabels()[5].set_color('black') 
    plt.gca().get_yticklabels()[8].set_color('black') 
    plt.gca().get_yticklabels()[9].set_color('black') 
    plt.gca().get_yticklabels()[10].set_color('black') 
    plt.gca().get_yticklabels()[11].set_color('black') 
    plt.gca().get_yticklabels()[12].set_color('black') 
    plt.gca().get_yticklabels()[15].set_color('black') 
    plt.gca().get_yticklabels()[16].set_color('black') 
    plt.gca().get_yticklabels()[17].set_color('black') 
    plt.gca().get_yticklabels()[20].set_color('black') 
    
    f.set_size_inches(width, height)
    f.savefig(path)
```

```python
def saveplot_lines(df, title, ylabel, path):
    from pathlib import Path
    import matplotlib as mpl
    import matplotlib.font_manager as fm
    import matplotlib.pyplot as plt
    mpl.use('pdf')

    fpath = Path(mpl.get_data_path(), "fonts/ttf/cmu_serif_roman.ttf")
    fprop = fm.FontProperties(fname=fpath)

    width = 6.2 #inches
    height = width

    font = {'family':'serif', 'serif': ['cmr10']}
    plt.rc('font',**font)
    plt.rcParams['axes.unicode_minus'] = False
    plt.rc('axes', titlesize=12)
    plt.rc('xtick', labelsize=12)
    plt.rc('ytick', labelsize=12)

    df_to_plot = df

    sns.set_theme(style="whitegrid", palette="pastel")
    f, ax = plt.subplots(figsize=(9, 8))
    f.subplots_adjust(left=0.14, bottom=.15, right=.99, top=.93)
    sns.pointplot(ax=ax,
            data=df_to_plot, x='party', y='value', hue='Year',
            markers=["^", "o"], 
            linestyles=["-", "--"],
            #kind="point",
            )
    
    ax.set_xlabel('Party', fontsize=12, font=fpath, color='black')
    ax.set_ylabel(ylabel, fontsize=12, font=fpath, color='black')
    ax.set_title(title, fontsize=12, font=fpath, color='black')
    ax.set_xticklabels(['Linke', 'Grüne', 'SPD', 'FDP', 'CDU/CSU', 'AfD'], size=12, font=fpath, color='black')
    #ax.set_yticklabels(ax.get_yticklabels(), size=12, font=fpath, color='black')
    for label in ax.get_yticklabels():
        label.set_fontproperties(fprop)
    plt.legend(prop=fprop)
    ax.get_legend().remove()
    f.set_size_inches(width, height*12/21)
    f.savefig(path)
```
